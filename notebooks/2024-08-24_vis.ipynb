```python
import os
import configparser
import sys

# 設定の読み込み
config = configparser.ConfigParser()
config_path = "../config.ini"
config.read(config_path)
PROJECT_DIR = config["paths"]["project_dir"]
sys.path.append(PROJECT_DIR)

from src.utils_experiment import OptimLogParser
```

```python
LOG_DIR = config["paths"]["logs_dir"]
res_files = sorted(os.listdir(LOG_DIR))
res_files
```

```python
parser = OptimLogParser(os.path.join(LOG_DIR, res_files[0]))
parser.parse_log_file()

print("Experimental settings: ")
print(parser.settings)
print()

print("Initial data: ")
display(parser.initial_data)
print()

print("BO data: ")
display(parser.bo_data)
print()

print(f"Best value: {parser.bo_data["function_value"].max()}")
```
