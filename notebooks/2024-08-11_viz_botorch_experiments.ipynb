```python
import os
import sys
import configparser

# 設定の読み込み
config = configparser.ConfigParser()
config_path = '../config.ini'
config.read(config_path)
PROJECT_DIR = config['paths']['project_dir']
EXPT_RESULT_DIR = config['paths']['results_dir']
LOG_DIR = config['paths']['logs_dir']
sys.path.append(PROJECT_DIR)

print(EXPT_RESULT_DIR)

import pandas as pd
```

```python
import optuna
import csv
import os

def load_csv_to_study(csv_file_path, study_name):
    study = optuna.create_study(study_name=study_name, direction="maximize")
    
    with open(csv_file_path, 'r') as csv_file:
        reader = csv.DictReader(csv_file)
        for row in reader:
            trial = study.ask()  # Create a new trial

            # ベストパラメータを取得
            best_x = eval(row["best_x"])
            # 目的関数の値を取得
            best_f = float(row["best_f"])
            
            # トライアルの報告前にユーザー属性を設定
            trial.set_user_attr("params", best_x)
            study.tell(trial, best_f)
    
    return study

def plot_history_from_file(csv_file_path):
    study_name = os.path.basename(csv_file_path).split('.')[0]  # ファイル名をStudy名に使用
    study = load_csv_to_study(csv_file_path, study_name)
    optuna.visualization.plot_optimization_history(study).show()
```

```python
file_path = os.path.join(EXPT_RESULT_DIR, '2024-08-11/botorch/griewank_3d/base', 'experiment_results_2024-08-11_21-00-26.csv')
plot_history_from_file(file_path)
```
