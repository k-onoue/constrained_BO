```python
# Initialize bounds for the optimization
bounds = torch.tensor([[-2.], [10.]])
    
# Generate initial samples
X_init = torch.Tensor([[-2], [10]])
Y_init = - test_function(X_init)

# Create an instance of DiscreteBO and initialize it
bo = DiscreteBO(bounds=bounds)
bo.initialize(X_init, Y_init)
    
# Optimization loop
for i in range(5):
    new_X = bo.step()
    new_Y = - test_function(new_X)
    bo.update(new_X, new_Y) 
    print(f"Iteration {i+1}: Suggested point: {new_X.numpy()}, Function value: {new_Y.numpy()}")

print()
print(f"Best value found: {bo.train_Y.min().item()}")
print(f"Best point found: {bo.train_X[bo.train_Y.argmin()].numpy()}")
```

```python
#!/usr/bin/env python3
# coding: utf-8

import os
import torch
from botorch.models import SingleTaskGP
from botorch.fit import fit_gpytorch_mll
from botorch.optim import optimize_acqf
from botorch.acquisition import ExpectedImprovement
from gpytorch.mlls.exact_marginal_log_likelihood import ExactMarginalLogLikelihood
import matplotlib.pyplot as plt

# 
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
dtype = torch.double

# 2
def objective_function(X):
    return -((X - 2) ** 2).sum(dim=-1)

# 
def generate_initial_data(n=5):
    train_x = torch.rand(n, 1, device=device, dtype=dtype) * 4  # [0, 4]
    train_y = objective_function(train_x).unsqueeze(-1)  # 
    return train_x, train_y

# 
def initialize_model(train_x, train_y):
    model = SingleTaskGP(train_x, train_y)
    mll = ExactMarginalLogLikelihood(model.likelihood, model)
    return mll, model

# 
def optimize_acqf_and_get_observation(acq_func):
    candidates, _ = optimize_acqf(
        acq_function=acq_func,
        bounds=bounds,
        q=1,
        num_restarts=10,
        raw_samples=20,
    )
    new_x = candidates.detach()
    new_y = objective_function(new_x).unsqueeze(-1)
    return new_x, new_y

# 
bounds = torch.tensor([[0.0], [4.0]], device=device, dtype=dtype)

# 
train_x, train_y = generate_initial_data(n=5)
mll, model = initialize_model(train_x, train_y)

# 
N_BATCH = 15  # 
for iteration in range(1, N_BATCH + 1):
    # 
    fit_gpytorch_mll(mll)
    
    # EI
    EI = ExpectedImprovement(model=model, best_f=train_y.max())
    
    # 
    new_x, new_y = optimize_acqf_and_get_observation(EI)
    
    # 
    train_x = torch.cat([train_x, new_x])
    train_y = torch.cat([train_y, new_y])
    
    # 
    mll, model = initialize_model(train_x, train_y)
```
